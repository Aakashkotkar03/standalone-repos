```python
# Define custom aggregation functions
agg_funcs = {
    'totalUsers': np.sum,
    'screenPageViews': np.sum,
    'sessions': np.sum,
    'eventCount': np.sum,
    'averageSessionDuration':np.mean
    # Use np.mean for this column
}

# Create a pivot table with custom aggregation functions
table = pd.pivot_table(result, values=['totalUsers', 'screenPageViews', 'sessions', 'eventCount','averageSessionDuration'],
                        index=['category'], aggfunc=agg_funcs)


table = table.sort_values(by='totalUsers', ascending=False)
table['% Users']=(table['totalUsers'])/(table['totalUsers'].sum())*100

# Print the resulting pivot table
print(table)
```

```python
# Create a DataFrame
pivot_table = pd.DataFrame(table)

# Export the DataFrame to an Excel file
pivot_table.to_excel('English_pivot_table.xlsx', index=True)
```
